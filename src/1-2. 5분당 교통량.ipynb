# 함수로 반복 돌리기

```python
month='09'
```

```python
# 파일 생성
head_df=pd.DataFrame(columns=['집계일자','집계시분','콘존ID','교통량','평균속도','통행시간','통행시간최대값','통행시간중위값','통행시간최소값',
         '출발노드이름','도착노드이름'])

head_df.to_csv(f'data/5분당_교통량(등교){month}월.csv', index=False)
```

```python
def read_traffic(date):
    df=pd.read_csv(f'data/2022{month}/2022{month}{date}.csv')
    df.columns=['집계일자','집계시분','콘존ID','차로유형구분코드','교통량','평균속도','통행시간','정상건수',
    '보정건수','통행시간표준편차','통행시간변동계수','통행시간최대값','통행시간중위값','통행시간최소값',
    '퍼센타일15_85통행시간','최종변경시각','출발노드이름','출발노드위도','출발노드경도','도착노드이름','도착노드위도','도착노드경도']
    
    df2=df[['집계일자','집계시분','콘존ID','교통량','평균속도','통행시간','통행시간최대값','통행시간중위값','통행시간최소값',
         '출발노드이름','도착노드이름']]
    
    df2=df2[df2['출발노드이름'].isin(['하남JC','하남IC','동서울TG','산곡JC','경기광주IC'])&
            df['콘존ID'].isin(['0352CZS400','0352CZS410','0352CZS420','0352CZS430',])]
    
    # 기존 CSV 파일 불러오기
    csv_file_path =f'data/5분당_교통량(등교){month}월.csv'

    # 데이터프레임을 CSV 파일에 이어서 쓰기
    df2.to_csv(csv_file_path, mode='a', header=False, index=False)

    print("이어쓰기가 완료되었습니다.")

    
    
```

```python
#다운받는 사이트에서 누락된 데이터들이 있음
#데이터가 없는 것은 error가 나기 때문에 try except로 함
print(month,'월')

for date in range(1,32):
    if date<10:
        date='0'+str(date)
    try:
        read_traffic(date)
    except:
        print(date,'오류가 발생했습니다')


```

```python
csv_file_path='data/5분당_교통량(등교)통합.csv'
for i in range(1,13):
    if i <10:
        i='0'+str(i)
    df3 =pd.read_csv(f'data/5분당_교통량(등교){i}월.csv')
    df3.to_csv(csv_file_path, mode='a', header=False, index=False)

    print("이어쓰기가 완료되었습니다.")
```

```python
df4=pd.read_csv('data/5분당_교통량(등교)통합.csv')
df4
```

```python
#preprocessing ,원하는 형식으로 가공
df4['집계일자'] = pd.to_datetime(df4['집계일자'], format='%Y%m%d')
df4['요일'] = df4['집계일자'].dt.day_name()
df4['요일'][df4['요일']=='Monday']='월'
df4['요일'][df4['요일']=='Tuesday']='화'
df4['요일'][df4['요일']=='Wednesday']='수'
df4['요일'][df4['요일']=='Thursday']='목'
df4['요일'][df4['요일']=='Friday']='금'
df4['요일'][df4['요일']=='Saturday']='토'
df4['요일'][df4['요일']=='Sunday']='일'
df4=df4[df4['출발노드이름']!='하남JC']
df4
```

```python
#다시 저장
head_df=pd.DataFrame(columns=['집계일자','집계시분','콘존ID','교통량','평균속도','통행시간','통행시간최대값','통행시간중위값','통행시간최소값',
         '출발노드이름','도착노드이름','요일'])

head_df.to_csv(f'data/5분당_교통량(등교)통합.csv', index=False)
df4.to_csv(csv_file_path, mode='a', header=False, index=False)
```
